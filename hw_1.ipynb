```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#1 y(x) = cos(k*x)
k1,k2 = 1,3.3

x = np.linspace(0,10,201)
y1,y2 = np.cos(k1*x),np.cos(k2*x)
plt.plot(x,y1)
plt.plot(x,y2)


#2 вычисление длины вектора:
def mod_vec(x,y,z):
    return np.sqrt(x*x+y*y+z*z)

print(mod_vec(2,3,4))
```

## Урок 2

```python
#### график окружности, эллипса, гиперболы
R,a,b,x0,y0 = 1,1,0.3, -1.2,3.1
x = np.linspace(x0-R+0.0001,x0+R-0.0001,20001)
y1 = y0 + np.sqrt(R*R - (x-x0)*(x-x0))
y2 = y0 - np.sqrt(R*R - (x-x0)*(x-x0))

plt.subplot(1,3,1)
plt.plot(x,y1,'b')
plt.plot(x,y2,'b')
plt.axis('equal')

plt.subplot(1,3,2)

x3 = np.linspace(-a+0.0001,a-0.0001,21)
y3 = np.sqrt(b*b - (b*x3/a)*(b*x3/a))
plt.plot(x3, y3,'r')
plt.plot(x3,-y3,'r')
plt.axis('equal')

plt.subplot(1,3,3)

x4 = np.linspace(-a+0.0001,a-0.0001,21)
y4 = np.sqrt((b*b + (b*x4/a)*(b*x4/a)))
plt.plot(x4, y4,'r')
plt.plot(x4,-y4,'r')
plt.axis('equal')
```

```python
# окружность в полярных кординатах
s=np.linspace(0,6.2831,2001)
x=np.sin(s)
y=np.cos(s)
plt.plot(x,y)
plt.axis('equal')
```
