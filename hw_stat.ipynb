```python
# S.1 Задача Бюффона: 
#- Иголку длиной b=10 см наудачу бросают на разлинованную бумагу, где расстояние между соседними линиями тоже a=10 см.
#- С какой вероятностью упавшая иголка пересечет линию бумаги?
#- Сделать расчеты, взяв другие (неравные значения a и b)
a,b1,b2,b3,N = 1,1,2.0,1000,1000000
Xb = a*np.random.rand(N) # координата нижней части иглы
Fi = np.pi*0.5*np.random.rand(N)
Xt1 = Xb+b1*np.sin(Fi)
Xt2 = Xb+b2*np.sin(Fi)
Xt3 = Xb+b3*np.sin(Fi)
n_t1,n_t2,n_t3 = 0,0,0
for i in range(N):
    if Xt1[i]>=a: n_t1 +=1
    if Xt2[i]>=a: n_t2 +=1
    if Xt3[i]>=a: n_t3 +=1
#    
print(a,b1,n_t1/N)  # ширина линеек равна длине иглы
print(a,b2,n_t2/N)  # ширина линеек вдвое длиннее иглы
print(a,b3,n_t3/N)  # ширина линеек в 1000 раз длиннее иглы
```

```python
# В плафоне освещения лифта -  три лампы. Вероятность перегорания каждой лампы 
# р = 0.1% в день. Ровно раз в месяц (30 дней) лифт посещает электрик и меняет все перегоревшие лампы на исправные. 
#
# Задача: Какова вероятность Р того, что (в какой-то момент между визитами монтера) все три лампы перегорят,
# и лифт останется без света?

# comment: вообще-то решение аналитическое должно быть таким  1-р0 -р1 -р2, 
# где р0 - вероятность бином.распределения выпадания ровно 0 событий (сгорания одной лампы) за 30 дней
#     р1 - вероятность бином.распределения выпадания ровно 1 событий (сгорания одной лампы) за 30 дней
#     р2 - вероятность бином.распределения выпадания ровно 2 событий (сгорания одной лампы) за 30 дней
#     pn = C^n_30 *(p^n)*(1-p)^(30-n)
```

```python
p=0.001
p0 = 1*p**0*(1-p)**30
p1 = 30*p**1*(1-p)**29
p2 = 30*29*0.5*p**2*(1-p)**28
print(p0,p1,p2)
1.0 - p0 - p1 - p2
```

```python
# Усложнение: если вдруг все три лампы перегорели, то жильцы звонят в сервис, и на следующий день 
# приходит электрик и меняет все лампы (т.е. примерно сутки лифт без света). 
# Найти коэффициент готовности Кг, т.е. средний % времени, 
# в течение которого в лифте горит хотя бы одна лампа (или наоборот, 1-Кг = % темного времени).
p3 = 1.0 /(1-p0-p1-p2) # один раз в такое количество месяцев происходит исследуемый случай и люди сидят днь без света
# стало быть коэфф неготовности будет 1 к p3*30
K_n = 1.0 /(30*p3)
print(K_n)
```
